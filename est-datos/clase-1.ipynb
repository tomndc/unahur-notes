## Definición básica

Vamos a pensar en términos de $Entrada \to Procesamiento\ algoritmico \to Salida$.

Un algoritmo es un conjunto finito de pasos para resolver un problema, con un lenguaje de programación, siendo este un conjunto de reglas y símbolos que permiten escribir instrucciones para que la computadora ejecute tareas.

- Variables
- Palabras reservadas
- Operadores
- Expresiones
- Entrada / Salida
- Comentarios
- Estructura de programa

Como elementos básicos, tenemos las variables:

- Son contenedores de datos, en una posición de memoria.
- Ocupan espacio en RAM.
- Tienen un nombre y un tipo, tiene un alias para la posición.
- Su tipo define el espacio de memoria ocupado, también define su codificación (como interpreta los 0s y 1s en la memoria).

Asociado a sus tipos, tienen tipos de datos primitivos:

- Ya están implementados en el lenguaje.
- Representación como datos numéricos, lógicos, etc.

### Númericos

Los enteros:

- De tipo `int`, números enteros con signo.
- De tipo `float`, números reales con decimales, tienen determinada precisión según su cantidad de cifras decimales.

### Strings

Son series de simbolos.

- Letras, números o especiales (ej. espacios)
- Palabras
- Se definen entre comillas (dobles o simples)

### Booleanos

El tipo `bool`

De lógica booleana.

- Verdadero (True)
- Falso (False)

Como usamos Python, es de tipado automático y dinámico, interpreta el tipo de dato.

Se puede cambiar el tipo de variable (casteo):
```python
int() # Entero
float() # Flotante
str() # String
```


### Operadores
| Operador       | Significado           |
|---------------|----------------------|
| `=`           | Asignación            |
| `+=`          | Suma y asignación     |
| `-=`          | Resta y asignación    |
| `+`           | Suma                 |
| `-`           | Resta                |
| `*`           | Multiplicación        |
| `/`           | División             |
| `%`           | Módulo (resto)        |
| `==`          | Igualdad              |
| `!=`          | Diferente             |
| `>`           | Mayor que             |
| `<`           | Menor que             |
| `>=`          | Mayor o igual que     |
| `<=`          | Menor o igual que     |

````python
# Operadores de Asignación
x = 5       # Asigna el valor 5 a x
x += 3      # Equivale a x = x + 3, ahora x es 8
x -= 2      # Equivale a x = x - 2, ahora x es 6

# Operadores Aritméticos
suma = 5 + 3         # Resultado: 8
resta = 10 - 4       # Resultado: 6
multiplicacion = 6 * 7  # Resultado: 42
division = 10 / 2    # Resultado: 5.0
modulo = 10 % 3      # Resultado: 1 (resto de la división)

# Operadores de Comparación
es_igual = (5 == 5)      # True
es_diferente = (5 != 3)  # True
mayor = (10 > 5)        # True
menor = (3 < 7)         # True
mayor_igual = (8 >= 8)  # True
menor_igual = (4 <= 6)  # True

# Imprimir resultados
print("Suma:", suma)
print("Resta:", resta)
print("Multiplicación:", multiplicacion)
print("División:", division)
print("Módulo:", modulo)

print("¿5 es igual a 5?", es_igual)
print("¿5 es diferente de 3?", es_diferente)
print("¿10 es mayor que 5?", mayor)
print("¿3 es menor que 7?", menor)
print("¿8 es mayor o igual que 8?", mayor_igual)
print("¿4 es menor o igual que 6?", menor_igual)
````

### Estructuras de control

Permiten modificar el orden de ejecución del programa.

Sus tipos son:

- Selectivas (`if, if-else, if-elif-else`).
- Repetitivas (`for, while`).

Las selectivas son bifurcaciones en el flujo del programa, decide que hacer.

````python
# Ejemplo de for

for var in range(10):
    print(var)
````

`limInf < numEntrada < LimSup` Python permite intervalos como en matemática.

En vez de un `and`: 

